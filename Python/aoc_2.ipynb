```python
raw_demo = """
    forward 5
    down 5
    forward 8
    up 3
    down 8
    forward 2
"""

def get_inputs(raw_str : str):
    intermediate = (x.replace("    ", "").split() for x in raw_str.splitlines())
    return [x for x in intermediate if x]

def get_new_state(current_state, an_input):
    command, value = an_input
    value = int(value)
    updater = {
        "forward": (lambda x, y : (x + value, y)),
        "up": (lambda x, y : (x, y - value)),
        "down": (lambda x, y : (x, y + value))
    }
    return updater[command](*current_state)

import unittest

class BaseTests(unittest.TestCase):

    def test_get_inputs(self):
        inputs = get_inputs(raw_demo)
        self.assertEquals(
            inputs,
            [
                ["forward", "5"],
                ["down", "5"],
                ["forward", "8"],
                ["up", "3"],
                ["down", "8"],
                ["forward", "2"]
            ],
            f"inputs = {inputs}"
        )

        state = (0, 0)

        state = get_new_state(state, inputs[0])
        self.assertEqual(state, (5, 0))

        state = get_new_state(state, inputs[1])
        self.assertEqual(state, (5, 5))

        state = get_new_state(state, inputs[2])
        self.assertEqual(state, (13, 5))

        state = get_new_state(state, inputs[3])
        self.assertEqual(state, (13, 2))

        state = get_new_state(state, inputs[4])
        self.assertEqual(state, (13, 10))

        state = get_new_state(state, inputs[5])
        self.assertEqual(state, (15, 10))

unittest.main(argv=[''], verbosity=2, exit=False)
```
