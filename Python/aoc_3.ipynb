```python
sample_input = """00100
11110
10110
10111
10101
01111
00111
11100
10000
11001
00010
01010
"""

from collections import Counter

def get_columns(lines):
    columns = []
    for i in range(0, len(lines[0])):
        col = []
        for line in lines:
            col.append(line[i])
        columns.append(col)
    return ["".join(x) for x in columns]

def sanitize(raw_input: str):
    raw_input = raw_input.splitlines()
    return get_columns(raw_input) , raw_input

def get_gamma_and_epsilon(raw_input : str):
    columns, _ = sanitize(raw_input)
    sample_data = [Counter(x) for x in columns]
    gamma_raw = ""
    epsilon_raw = ""
    for column in sample_data:
        sorted_cols = column.most_common()
        gamma_raw = gamma_raw + sorted_cols[0][0]
        epsilon_raw = epsilon_raw + sorted_cols[1][0]
    return int(gamma_raw, 2), int(epsilon_raw, 2)

def get_power_consumption(raw_input: str):
    gamma, epsilon = get_gamma_and_epsilon(raw_input)
    return gamma * epsilon

assert get_gamma_and_epsilon(sample_input) == (22, 9)
assert get_power_consumption(sample_input) == 198
```

```python
def get_value(raw_input: str, predicate):
    columns, lines = sanitize(raw_input)
    filtered = list(lines)
    for index in range(0, len(lines[0])):
        columns = get_columns(filtered)
        counter = Counter(columns[index])
        lookup = predicate(counter)
        filtered = [x for x in filtered if x[index] == lookup]
        if len(filtered) == 1:
            return int(filtered[0], 2)

def get_oxygen_generator_predicate(item : Counter):
    a_sorted = item.most_common()
    if len(a_sorted) > 1:
        if a_sorted[0][1] == a_sorted[1][1]:
            return '1'
    return a_sorted[0][0]

def get_co2_scrubber_rating(item : Counter):
    a_sorted = item.most_common()
    if len(a_sorted) > 1:
        if a_sorted[0][1] == a_sorted[1][1]:
            return '0'
    if len(a_sorted) == 1:
        return a_sorted[0][0]
    return a_sorted[1][0]

assert get_value(sample_input, get_oxygen_generator_predicate) == 23
assert get_value(sample_input, get_co2_scrubber_rating) == 10

def get_life_support_rating(raw_input: str):
    return get_value(raw_input, get_oxygen_generator_predicate) * get_value(raw_input, get_co2_scrubber_rating)

assert get_life_support_rating(sample_input) == 230
```
